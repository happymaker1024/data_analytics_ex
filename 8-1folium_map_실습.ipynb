# 지도에 좌표 표시하기
- 공식문서 : https://python-visualization.github.io/folium/latest/getting_started.html#Choropleth-maps

```python
import folium
```

```python
# 디폴트 위경도 지정(포커싱), 지도사이트 Map 객체 생성
folium_map = folium.Map(location=[37.5662952, 126.9779451], zoom_start=16)
```

```python
folium.Marker(location=[37.5662952, 126.9779451]).add_to(folium_map)
```

```python
popup = folium.Popup('서울시청 위치', max_width=300)
folium.Marker(location=[37.5662952, 126.9779451], popup=popup).add_to(folium_map)
```

```python
import os

# 존재하지 않으면 폴더를 만듬
os.makedirs('map_data', exist_ok=True)

folium_map.save('map_data/folium_map.html')
```
