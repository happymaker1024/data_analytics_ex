# 서울시 열린데이터 광장

```python
from dotenv import load_dotenv
import os
from pathlib import Path

# .env 파일에 예: SEOUL_OPEN_KEY=your_key

env_path = Path('.') / '.env'
load_dotenv(dotenv_path=env_path)

SEOUL_OPEN_KEY = os.getenv("SEOUL_OPEN_KEY")
if SEOUL_OPEN_KEY is None:
    raise RuntimeError("SEOUL_OPEN_KEY가 .env에 없습니다.")
# SEOUL_OPEN_KEY
```

# json 타입으로 데이터 불러오기

# xml만 지원할 경우
- xml -> json -> dataframe으로 변환
- pip install xmltodict

```python
import pandas as pd
import requests
import xmltodict

# 1. url 파라미터 정의
serviceKey = SEOUL_OPEN_KEY
service_id = "bikeList"
data_type = "xml"
date = "20251001"
count = 10

url = f'http://openapi.seoul.go.kr:8088/{serviceKey}/{data_type}/{service_id}/1/{count}'


# 2. 요청하기
response = requests.get(url).text

# 3. XML → dict
data = xmltodict.parse(response)
data

# 4. 데이터 추출
rows = data["rentBikeStatus"]["rentBikeStatus"]["row"]
rows

# 5. 데이터프레임으로 변환 및 확인
df = pd.DataFrame(rows)
print(df.shape)
df.head()
```
