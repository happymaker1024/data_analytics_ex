```python
import pandas as pd
import requests
import xmltodict

url = f"https://kosmes.or.kr/opendata/portal/openapi/MTRPTEQMTINDST"
response = requests.get(url)
# print(response.text)
```

```python
# 3. XML → dict
data = xmltodict.parse(response.text)
data['MTRPTEQMTINDST']['row']
```

```python
df = pd.DataFrame(data['MTRPTEQMTINDST']['row'])
df.head()
```
