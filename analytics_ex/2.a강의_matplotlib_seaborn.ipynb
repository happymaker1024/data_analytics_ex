```python
data_path = 'datas/fuel_econ.csv'
df = pd.read_csv(data_path)
df.head()
```

컬럼명	의미	설명
- id	차량 고유 식별자	데이터셋 내에서 각 차량을 구분하기 위한 고유 ID입니다.
- make	제조사	자동차 브랜드명 (예: Toyota, Ford, BMW 등)
- model	모델명	차량의 모델명 (예: Camry, Accord, Model 3 등)
- year	연식	해당 차량의 제조 연도
- VClass	차량 등급(Vehicle Class)	차량의 분류 (예: Compact Car, SUV, Pickup 등)
- drive	구동 방식	자동차의 구동 형태 (예: FWD=전륜, RWD=후륜, AWD=4륜구동 등)
- trans	변속기	변속기의 종류 (예: Auto=자동, Manual=수동, CVT 등)
- fuelType	연료 종류	차량이 사용하는 연료 (예: Gasoline, Diesel, Electric 등)
- cylinders	실린더 수	엔진의 실린더 개수 (예: 4, 6, 8 등)
- displ	배기량(Displacement)	엔진의 배기량(Liter 단위) — 예: 2.0은 2리터급 엔진
- pv2	2-사이클 엔진 연비 보정계수	EPA의 측정 과정에서 사용되는 보정값(Primary vehicle 2-cycle test)
- pv4	4-사이클 엔진 연비 보정계수	4-cycle 보정 테스트 수치(보정된 시험값으로 실제 주행과의 차이 보정)
- city	도심 연비 (MPG)	시내 주행 기준 공식 연비 (Miles per Gallon)
- UCity	도심 실제연비(보정 후)	실제 운전 환경을 반영한 도심 주행 연비
- highway	고속도로 연비 (MPG)	고속도로 주행 기준 공식 연비
- UHighway	고속도로 실제연비(보정 후)	실제 주행 환경을 반영한 고속도로 연비
- comb	복합 연비 (MPG)	도심과 고속도로 연비를 평균한 복합 연비 (일반적으로 city 55% + highway 45%)
- co2	이산화탄소 배출량 (g/mile)	주행 거리 1마일당 배출되는 CO₂ 양 (단위: g/mile)
- feScore	연비 점수 (Fuel Economy Score)	EPA가 부여하는 연비 등급 점수 (보통 1~10 사이, 높을수록 효율적)
- ghgScore	온실가스 점수 (Greenhouse Gas Score)	온실가스 배출 효율 점수 (높을수록 친환경적)

## 문제 2. 엔진 크기와 이산화탄소 배출량의 관계를 분석하세요. 
- (히트맵으로)   displ 이 엔진사이즈이고, co2 가 이산화탄소 배출량입니다.  트랜드를 분석하세요

```python
#  엔진 크기와 이산화탄소 배출량의 관계 분석하기
# (히트맵으로)   displ 이 엔진사이즈이고, co2 가 이산화탄소 배출량입니다.  트랜드 분석하기
sb.heatmap(
    data= df[['displ','co2']].corr(), 
    annot= True , 
    fmt= '.2f', 
    cmap= 'coolwarm')
plt.title('엔진 크기와 이산화탄소 배출량의 관계')
```
