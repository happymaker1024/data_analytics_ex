- 리뷰점수가 95점 이상이고, Australia 또는 New Zealand 에서 만들어진 와인 데이터를 가져오세요

```python
(df['points'] >=95)
```

```python
((df['country'] == 'Australia') | (df['country'] == 'New Zealand'))
```

```python
filter =(df['points'] >=95) & ((df['country'] == 'Australia') | (df['country'] == 'New Zealand'))
```

```python
# 리뷰점수가 95점 이상이고, Australia 또는 New Zealand 에서 만들어진 와인 데이터를 가져오세요
# df.loc[행, 열]
df.loc[filter , : ].head()
```
